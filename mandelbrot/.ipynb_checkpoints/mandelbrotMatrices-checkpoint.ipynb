```python
#
# Mandelbrot/Julia sets plot using complex matrices.
#

import numpy as np
import matplotlib.pyplot as plt

d, n = 100, 50  # pixel density & number of iterations
r = 2.5  # escape radius (must be greater than 2)

# Mandelbrot
#x = np.linspace(-2.5, 1.5, 4 * d + 1)
#y = np.linspace(-1.5, 1.5, 3 * d + 1)

# Julia
x = np.linspace(-2, 2, 3 * d + 1)
y = np.linspace(-2, 2, 3 * d + 1)

A, B = np.meshgrid(x, y)

# Mandelbrot
#C = A + B * 1j

# Julia
Z = A + B * 1j

# Mandelbrot
#Z = np.zeros_like(C)

# Julia
C = np.zeros_like(Z)
C = C + complex('(-0.8+0.156j)')

# Mandelbrot
#T = np.zeros(C.shape)

# Julia
T = np.zeros(Z.shape)

for k in range(n):
    M = abs(Z) < r
    Z[M] = Z[M] ** 2 + C[M]
    T[M] = k + 1

plt.imshow(T, cmap=plt.cm.twilight_shifted)
# plt.savefig("mandelbrot.png", dpi=200)
```
